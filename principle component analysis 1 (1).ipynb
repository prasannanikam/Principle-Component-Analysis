```python
#importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
df= pd.read_csv("wine.csv")
```

```python
df.head()
```

```python
df["Type"].value_counts()
```

```python
data=df.iloc[:,1:]
```

```python
data.head()
```

```python
data.info()
```

```python
dataar= data.values
dataar
```

```python
norm=scale(dataar)
norm
```

```python
pca = PCA()
pca_values= pca.fit_transform(norm)
pca_values
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
```

```python
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color="aqua")
```

```python
data1=pd.concat([df["Type"],pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3'])],axis=1)
```

```python
data1.head()
```

```python
fig = plt.figure(figsize=(16,12))
sns.scatterplot(data=data1)
```

Checking with other Clustering Algorithms


```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
## Creating dendograms

plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(norm,'complete'))
```

```python
# Creating clusters 
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(norm), columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# adding clusters to data set

winedata3 = df.copy()
winedata3['clustersid']=hclusters.labels_
winedata3
```

K-Means Clustering


```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,6):
    kmeans = KMeans(n_clusters=i, random_state=2)
    kmeans.fit(norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot Elbow graph

plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python

clusters3 = KMeans(3,random_state=30).fit(norm)
clusters3
```

```python
clusters3.labels_
```

```python
winedata4 =df.copy()
winedata4['clusters3id']=clusters3.labels_
winedata4
```

```python
winedata4['clusters3id'].value_counts()
```

```python
# Cluster algorithm using k=4

clusters4 = KMeans(4, random_state=30).fit(norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assigning clusters to data set

winedata5 =df.copy()
winedata5['clusters4id'] = clusters4.labels_
winedata5
```

```python
winedata5['clusters4id'].value_counts()
```
